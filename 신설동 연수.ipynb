```python
#2020      ( )
import csv
f=open('2020korea.csv')
data=csv.reader(f)
m=[]
f=[]
for row in data:
    if '' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3].replace(',','')))
            f.append(int(row[-(i+1)].replace(',','')))
f.reverse()

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(11,6),dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title('    ')
plt.barh(range(101),m, label='')
plt.barh(range(101),f,label='')
plt.legend()
plt.show()
```

```python
#    ()
import csv
f=open('2020.csv')
data=csv.reader(f)
ch=0
ad=0
se=0
size=[]

name=input('   :')

for row in data:
    if name in row[0]:
        for i in range(3,18):
            ch += int(row[i].replace(',',''))
        for i in range(18,68):
            ad += int(row[i].replace(',',''))
        for i in range(68,104):
            se += int(row[i].replace(',',''))
size.append(ch)
size.append(ad)
size.append(se)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font', family='Malgun Gothic')
color=['lime','royalblue','yellow']
plt.axis('equal')

plt.pie(size, labels=['','',''], autopct='%.1f%%', colors=color)
plt.title(name+'    ')
plt.show()
```

```python
#()   
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import csv
f=open('2020.csv')
data=csv.reader(f)
next(data)
data=list(data)

name=input('   : ')
mn=1
result_name=''
result=0

for row in data:
    if name in row[0]:
        for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
        home=np.array(row[3:],dtype=int)/int(row[2])
for row in data:
    for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
    away=np.array(row[3:],dtype=int)/int(row[2])
    s=np.sum(abs(home-away))
    if s<mn and name not in row[0]:
        mn=s
        result_name=row[0]
        result=away
                
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(11,6), dpi=300)
plt.rc('font',family='Malgun Gothic')
plt.title(name+'       ')
plt.plot(home, label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```

```python
#() 5     ?( )
import csv
f=open('elementary.csv')
data=csv.reader(f)
name=input('   :')
result=[]
gap=0

for row in data:
    if name in row[0]:
        for i in row[1:]:
            result.append(int(i.replace(',','')))
        gap=int(row[-1].replace(',',''))-int(row[1].replace(',',''))
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title(name+'   5  ')
plt.xticks(range(5),['2016','2017','2018','2019','2020'])
plt.xlabel('')
plt.ylabel('')
plt.plot(result)
plt.show()
print(' ():',gap)
```
