#7. (심화)우리 동네와 인구 구성이 비슷한 동네 찾아보기

```python
#7. (심화)우리 동네와 인구 구성이 비슷한 동네
import numpy as np
np.seterr(divide='ignore', invalid='ignore')                 
import csv
f=open('2020age.csv', encoding='cp949')
data=csv.reader(f)
next(data)                                                               
data=list(data)

name=input('궁금한 동네 이름 입력: ')
mn=1
result_name=''
result=0

for row in data:
    if name in row[0]:
        for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
        home=np.array(row[3:],dtype=int)/int(row[2])
for row in data:
    for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
    away=np.array(row[3:],dtype=int)/int(row[2])
    s=np.sum(abs(home-away))
    if s<mn and name not in row[0]:
        mn=s
        result_name=row[0]
        result=away
                
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(11,6), dpi=300)
plt.rc('font',family='NanumGothic')
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```

<2020년 전국 **시군구** 0세~100세 이상 연령별 인구 데이터 사용> -남녀구분X 총합

#8.(심화)인구 구성이 비슷한 지역(시군구) 찾아보기

```python
#8.(심화)인구 구성이 비슷한 지역구 찾아보기
import numpy as np
np.seterr(divide='ignore', invalid='ignore')
import csv
f=open('2020gu.csv', encoding='cp949')
data=csv.reader(f)
next(data)
data=list(data)

name=input('궁금한 시.군.구 이름을 입력하세요(예.종로구): ')  
mn=1
result_name=''
result=0

for row in data:
    if name in row[0]:
        for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
        home=np.array(row[3:],dtype=int)/int(row[2])
for row in data:
    for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
    away=np.array(row[3:],dtype=int)/int(row[2])
    s=np.sum(abs(home-away))
    if s<mn and name not in row[0]:
        mn=s
        result_name=row[0]
        result=away
                
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(11,6), dpi=300)
plt.rc('font',family='NanumGothic')
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```

#9.(심화)우리 동네(구)의 초등학생은 정말 감소했을까?(꺾은선 그래프)

```python
#9.(심화)우리 동네(구)의 초등학생은 정말 감소했을까?(꺾은선 그래프)
import csv
f=open('elementary.csv', encoding='cp949')
data=csv.reader(f)
name=input('알고싶은 구 이름을 입력하세요:')
result=[]
gap=0                                                                              #2020년과 2016년의 초등학생 인구수 차이 

for row in data:
    if name in row[0]:
        for i in row[1:]:
            result.append(int(i))
        gap=int(row[-1])-int(row[1])

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='NanumGothic')
plt.title(name+' 지역의 최근 5년간 초등학생수 변화')
plt.xticks(range(5),['2016','2017','2018','2019','2020'])    # x축 눈금
plt.xlabel('연도')
plt.ylabel('학생수')
plt.plot(result)                                                                  # 꺾은선 그래프 그리기 
plt.show()
print('학생수 차이(명):',gap)
```
