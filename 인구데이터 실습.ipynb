```python
#1. 2020년 전국 인구 피라미드 그래프 (교과서 업데이트)
import csv
f=open('202012korea.csv', encoding = 'cp949')
data = csv.reader(f)
m=[]                                                                          # 남성 데이터 변수 생성
f=[]                                                                            # 여성 데이터 변수 생성
for row in data:
    if '전국' in row[0]:                                                  # for 반복문으로 '전국' 연령별 데이터를 남/여 리스트에 붙임.
        for i in range(0,101):
            m.append(-int(row[i+3].replace(',','')))          # 남성 데이터는 음의 방향으로 
            f.append(int(row[-(i+1)].replace(',','')))
f.reverse()                                                                # 여성 데이터 인덱스를 끝에서부터 설정했기 때문에 0세부터로 되돌림. 

import matplotlib.pyplot as plt                                 
plt.style.use('ggplot')                                               # 배경 스타일 선택 (bmh, fast, dark_background, fivethirtyeight 등)
plt.figure(figsize=(11,6),dpi=300)                            # 그래프 크기와 해상도
plt.rc('font', family='NanumGothic')                         # 글꼴
plt.rcParams['axes.unicode_minus']=False            # 마이너스(-)부호 깨짐 방지
plt.title('전국 남녀 성별 인구 분포(2020년)')           # 그래프 제목
plt.barh(range(101),m, label='남성')                       # 막대 그래프 그리기 
plt.barh(range(101),f,label='여성')
plt.legend()                                                             # 범례 표시
plt.show()                                                               # 그래프 출력
```

```python
#6. 우리 동네의 고령화 정도 알아보기(연령별 인구 구성 원그래프 그려보기)
import csv
f=open('2020age.csv', encoding = 'cp949')
data=csv.reader(f)
ch=0                                                                     # 유소년층(Child) 변수 초기화
ad=0                                                                     # 청장년층(Adult) 변수 초기화
se=0                                                                     # 노년층(Senior) 변수 초기화
size=[]                                                                  # 원그래프 비율 크기를 담을 리스트 변수 생성

name=input('알고 싶은 동네를 입력해주세요:')

for row in data:
                                                                      # if 조건문으로 찾고 싶은 동네 선택 
        for i in range(3,18):
            ch += int(row[i].replace(',',''))
        for i in range(18,68):
            ad += int(row[i].replace(',',''))
        for i in range(68,104):
            se += int(row[i].replace(',',''))
size.append(ch)                                                 # size 리스트에 각 연령층별 인원수합계 붙이기 
size.append(ad)
size.append(se)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,5),dpi=300)
plt.rc('font', family='NanumGothic')
color=['lime','royalblue','yellow']                       # 각각의 색상 선택          
plt.axis('equal')                                                # x축, y축 길이 동일하게 

plt.pie(size, explode=(0,0,0.1), labels=['유소년층','청장년층','노년층'], autopct='%.1f%%', colors=color)   # 파이차트 그리기(크기, 강조, 레이블, 소수점자리, 색상)
plt.title(name+' 지역의 인구 구성 비율')        
plt.show()                                                     
print('*기준: 노년층 비율 7%이상 - 고령화 사회, 14%이상- 고령 사회, 20%이상- 초고령 사회')
```

#7. (심화)우리 동네와 인구 구성이 비슷한 동네 찾아보기

```python
#7. (심화)우리 동네와 인구 구성이 비슷한 동네
import numpy as np
np.seterr(divide='ignore', invalid='ignore')                 
import csv
f=open('2020age.csv', encoding='cp949')
data=csv.reader(f)
next(data)                                                               
data=list(data)

name=input('궁금한 동네 이름 입력: ')
mn=1
result_name=''
result=0

for row in data:
    if name in row[0]:
        for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
        home=np.array(row[3:],dtype=int)/int(row[2])
for row in data:
    for i in range(0,len(row[0:])):
            row[i]=row[i].replace(',','')
    away=np.array(row[3:],dtype=int)/int(row[2])
    s=np.sum(abs(home-away))
    if s<mn and name not in row[0]:
        mn=s
        result_name=row[0]
        result=away
                
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(11,6), dpi=300)
plt.rc('font',family='NanumGothic')
plt.title(name+' 지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label=name)
plt.plot(result,label=result_name)
plt.legend()
plt.show()
```

```python
#9.(심화)우리 동네(구)의 초등학생은 정말 감소했을까?(꺾은선 그래프)
import csv
f=open('elementary.csv', encoding='cp949')
data=csv.reader(f)
name=input('알고싶은 구 이름을 입력하세요:')
result=[]
gap=0                                                                              #2020년과 2016년의 초등학생 인구수 차이 

for row in data:
    if name in row[0]:
        for i in row[1:]:
            result.append(int(i.replace(',','')))
        gap=int(row[-1].replace(',',''))-int(row[1].replace(',',''))

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='NanumGothic')
plt.title(name+' 지역의 최근 5년간 초등학생수 변화')
plt.xticks(range(5),['2016','2017','2018','2019','2020'])    # x축 눈금
plt.xlabel('연도')
plt.ylabel('학생수')
plt.plot(result)                                                                  # 꺾은선 그래프 그리기 
plt.show()
print('학생수 차이(명):',gap)
```
