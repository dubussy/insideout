```python
# 선생님 제가 만들고 싶은 코드가 8세~ 13세의 인구가 가장 많은 도시 상위 10개 정도 그래프를 만들고 싶어요.
1. 만 나이 : 7세~12세
2. 도시 :  기준? 동? 시군구? 시도?
3. 인구가 가장 많다 : 절대 수치가 많은 것인가? 비율이 많은 것인가? 
4. 그래프 : 순위 비교라면 막대그래프? 
5. 분석방법: Pandas 사용? 일반? 
```

```python
# 기준: 전국 시군구, 절대 수치가 많은 지역 top10 막대그래프 그리기
import pandas as pd                                      #판다스 사용
df=pd.read_csv('20gu.csv',  encoding='cp949',
                          usecols =['행정구역','7세','8세','9세','10세','11세','12세'])            #지역명, 7~12세 column만 사용

df  #데이터프레임 보여주세요
```

```python
# 지역별 초등학생수 구하기
df['초등학생수'] = df['7세'] + df['8세'] + df['9세'] + df['10세'] + df['11세'] + df['12세']
df
```

```python
#초등학생수가 많은 지역 상위 10개 
df1=df.sort_values(by=['초등학생수'],ascending=False)   #내림차순으로 정렬
df1.head(10)                                                                      #상위 10개 지역만 보여주기
```

```python
df2=df1.head(10)     #위 데이터프레임을 df2란 이름으로 저장
```

```python
#막대그래프로 그리기 

import matplotlib.pyplot as plt

matplotlib.rcParams['font.family']='Malgun Gothic'
plt.figure(figsize=(11, 5), dpi=300)
plt.title('초등학생이 가장 많은 시 top10')
plt.xlabel('행정구역')
plt.ylabel('초등학생수')
plt.xticks(rotation=90)                                  #지역명이 길어서  세로로 회전
plt.bar(df2['행정구역'], df2['초등학생수'])    #df2 데이터로 (x축, y축 )설정
plt.show()
```
